```python
import os
from dotenv import load_dotenv
from tavily import TavilyClient

load_dotenv()

client = TavilyClient(api_key=os.getenv('TAVILY_API_KEY'))
```

```python
url = "https://blog.bytebytego.com/p/nextdoors-database-evolution-a-scaling"

response = client.extract(urls=[url])

for result in response.get('results', []):
    print(f"URL: {result['url']}")
    print(f"Text:\n{result['raw_content']}")
```

```python
import anthropic

claude = anthropic.Anthropic()

for result in response.get('results', []):
    raw = result['raw_content']
    url = result['url']

    print(raw)

    summary_response = claude.messages.create(
        model="claude-haiku-4-5-20251001",
        max_tokens=512,
        messages=[{
            "role": "user",
            "content": (
                "Write a concise one-paragraph summary of the following article:\n\n"
                f"{raw}"
            )
        }]
    )
    summary = summary_response.content[0].text

    print(f"=== URL: {url} ===\n")
    print("\n--- Summary ---")
    print(summary)
    print("\n" + "=" * 80 + "\n")
```
