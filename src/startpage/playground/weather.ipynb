```python
import python_weather

WEATHER_SYMBOL = {
    "UNKNOWN": "‚ú®",
    "CLOUDY": "‚òÅÔ∏è",
    "FOG": "üå´",
    "HEAVY_RAIN": "üåß",
    "HEAVY_SHOWERS": "üåß",
    "HEAVY_SNOW": "‚ùÑÔ∏è",
    "HEAVY_SNOW_SHOWERS": "‚ùÑÔ∏è",
    "LIGHT_RAIN": "üå¶",
    "LIGHT_SHOWERS": "üå¶",
    "LIGHT_SLEET": "üåß",
    "LIGHT_SLEET_SHOWERS": "üåß",
    "LIGHT_SNOW": "üå®",
    "LIGHT_SNOW_SHOWERS": "üå®",
    "PARTLY_CLOUDY": "‚õÖÔ∏è",
    "SUNNY": "‚òÄÔ∏è",
    "THUNDERY_HEAVY_RAIN": "üå©",
    "THUNDERY_SHOWERS": "‚õà",
    "THUNDERY_SNOW_SHOWERS": "‚õà",
    "VERY_CLOUDY": "‚òÅÔ∏è",
}

WIND_ARROWS = {
    "NORTH": "‚Üë",
    "NORTH_NORTHEAST": "‚Üë",
    "NORTHEAST": "‚Üó",
    "EAST_NORTHEAST": "‚Üó",
    "EAST": "‚Üí",
    "EAST_SOUTHEAST": "‚Üò",
    "SOUTHEAST": "‚Üò",
    "SOUTH_SOUTHEAST": "‚Üò",
    "SOUTH": "‚Üì",
    "SOUTH_SOUTHWEST": "‚Üô",
    "SOUTHWEST": "‚Üô",
    "WEST_SOUTHWEST": "‚Üô",
    "WEST": "‚Üê",
    "WEST_NORTHWEST": "‚Üñ",
    "NORTHWEST": "‚Üñ",
    "NORTH_NORTHWEST": "‚Üñ",
}

location = "Purmerend"


async def main() -> None:
    # Declare the client. The measuring unit used defaults to the metric system (celcius, km/h, etc.)
    async with python_weather.Client(unit=python_weather.METRIC) as client:
        # Fetch a weather forecast from Amsterdam.
        weather = await client.get(location)

        # Get today's forecast (first item in the daily_forecasts list)
        today = weather.daily_forecasts[0]

        # Get the weather symbol
        symbol = WEATHER_SYMBOL.get(weather.kind.name, "‚ùì")
        
        # Get wind arrow
        wind_arrow = WIND_ARROWS.get(weather.wind_direction.name, "?")
        
        # Format the output
        print(f"{weather.location}:")
        print(f"{symbol} {today.lowest_temperature}¬∞C - {today.highest_temperature}¬∞C")
        print(f"Humidity: {weather.humidity}% Precipitation: {weather.precipitation}mm")
        print(f"Wind: {wind_arrow} {weather.wind_speed} km/h")


await main()
```
